```python
import pandas as pd
import numpy as np
import finder
import paragraph
from typing import List
from unsupervisedpcfg.unsupervisedpcfg import UnsupervisedPCFG
```

```python
pp_nom = list(finder.parse_annotated(finder.read_files(['/home/piggy/lab/mycology/skol/tmp/nomenclatures.txt.ann'])))
pp_des = list(finder.parse_annotated(finder.read_files(['/home/piggy/lab/mycology/skol/tmp/descriptions.txt.ann'])))
```

```python
perm_nom = np.random.permutation(pp_nom)
perm_des = np.random.permutation(pp_des)
sub_nom = perm_nom[:100]
sub_des = perm_des[:100]
```

```python
paragraph.Paragraph.set_reinterpretations(['punctuation', 'year'])
def reint(corpus: List[paragraph.Paragraph]) -> List[str]:
    return [[w.lower() for w in p.reinterpret().split()] for p in corpus if p.reinterpret().split()]

vocab = np.unique([item.lower() for sublist in reint(np.concatenate([sub_nom, sub_des])) for item in sublist])
corpus_size = 50
test_size = 10
corpus_nom = reint(sub_nom[:corpus_size])
test_nom = reint(sub_nom[corpus_size:corpus_size + test_size])
test_des = reint(sub_des[:test_size])
```

```python
corpus_nom
```

```python
non_terminal_count = 10
m = UnsupervisedPCFG(non_terminal_count)
m.fit(corpus_nom, vocab)
```

```python
m.score(test_nom)
```

```python
m.score(test_des)
```
