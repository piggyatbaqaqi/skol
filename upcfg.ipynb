```python
import pandas as pd
import numpy as np
import finder
import paragraph
from typing import List
from unsupervisedpcfg.unsupervisedpcfg import UnsupervisedPCFG
```

```python
pp_nom = list(finder.parse_annotated(finder.read_files(['/home/piggy/lab/mycology/skol/tmp/nomenclatures.txt.ann'])))
pp_des = list(finder.parse_annotated(finder.read_files(['/home/piggy/lab/mycology/skol/tmp/descriptions.txt.ann'])))
pp_bib = list(finder.parse_annotated(finder.read_files(['/home/piggy/lab/mycology/skol/tmp/bibliography.txt.ann'])))
```

```python
perm_nom = np.random.permutation(pp_nom)
perm_des = np.random.permutation(pp_des)
perm_bib = np.random.permutation(pp_bib)
sample_size = 150
sub_nom = perm_nom[:sample_size]
sub_des = perm_des[:sample_size]
sub_bib = perm_bib[:sample_size]
```

```python
paragraph.Paragraph.set_reinterpretations(['punctuation', 'year'])
def reint(corpus: List[paragraph.Paragraph]) -> List[str]:
    return [[w.lower() for w in p.reinterpret().split()] for p in corpus if p.reinterpret().split()]

vocab = np.unique([item.lower() for sublist in reint(np.concatenate([sub_nom, sub_des, sub_bib])) for item in sublist])
corpus_size = 100
test_size = 20
corpus_nom = reint(sub_nom[:corpus_size])
test_nom = reint(sub_nom[corpus_size:corpus_size + test_size])
test_des = reint(sub_des[:test_size])
test_bib = reint(sub_bib[:test_size])
```

```python
corpus_nom
```

```python
non_terminal_count = 10
m = UnsupervisedPCFG(non_terminal_count)
m.fit(corpus_nom, vocab)
```

```python
m.score(test_nom)
```

Here we see that Description paragraphs have significantly worse scores than Nomenclature paragraphs.

```python
m.score(test_des)
```

Bibliography paragraphs are mostly worse, but there is some overlap in scores even on this small sample. Perhaps we can learn a grammar for Bibliogrphy paragraphs too, as they do have a roughly consistent form.

```python
m.score(test_bib)
```
