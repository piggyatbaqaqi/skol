```python
import os

import joblib
import numpy as np
import pandas as pd
from unsupervisedpcfg.unsupervisedpcfg import UnsupervisedPCFG
from typing import List

import finder
import paragraph
np.random.seed(1234)
```

```python
SRC_DIR='/home/piggy/lab/mycology/skol/src/github.com/piggyatbaqaqi/skol'
DATA_DIR=os.path.join(SRC_DIR, 'data/intermediate')
pp_nom = list(finder.parse_annotated(finder.read_files([os.path.join(DATA_DIR, 'nomenclatures.txt.ann')])))
pp_des = list(finder.parse_annotated(finder.read_files([os.path.join(DATA_DIR, 'descriptions.txt.ann')])))
pp_bib = list(finder.parse_annotated(finder.read_files([os.path.join(DATA_DIR, 'bibliography.txt.ann')])))
```

```python
sample_size = 1020
sub_nom = np.random.permutation(pp_nom)[:sample_size]
sub_des = np.random.permutation(pp_des)[:sample_size]
sub_bib = np.random.permutation(pp_bib)[:sample_size]
```

```python
paragraph.Paragraph.set_reinterpretations(['punctuation', 'year', 'latinate'])
def reint(corpus: List[paragraph.Paragraph]) -> List[str]:
    return [[w.lower() for w in p.reinterpret().split()] for p in corpus if p.reinterpret().split()]

vocab = np.unique([item.lower() for sublist in reint(np.concatenate([sub_nom, sub_des, sub_bib])) for item in sublist])
corpus_size = 200
test_size = 20
corpus_nom = reint(sub_nom[:corpus_size])
corpus_des = reint(sub_des[:corpus_size])
corpus_bib = reint(sub_bib[:corpus_size])
test_nom = reint(sub_nom[corpus_size:corpus_size + test_size])
test_des = reint(sub_des[:test_size])
test_bib = reint(sub_bib[corpus_size:corpus_size + test_size])
```

```python
non_terminal_count = 5
m_nom = UnsupervisedPCFG(non_terminal_count)
m_nom.fit(corpus_nom, vocab)
```

```python
joblib.dump(m_nom, os.path.join(SRC_DIR, 'results/models/m_nom.joblib'))
```

```python
m_nom_nom = pd.DataFrame(m_nom.score(test_nom)[2])
m_nom_nom
```

Here we see that Description paragraphs have significantly worse scores than Nomenclature paragraphs.

```python
m_nom_des = pd.DataFrame(m_nom.score(test_des)[2])
m_nom_des
```

Bibliography paragraphs are mostly worse, but there is some overlap in scores even on this small sample. Perhaps we can learn a grammar for Bibliography paragraphs too, as they do have a roughly consistent form.

```python
m_nom_bib = pd.DataFrame(m_nom.score(test_bib)[2])
m_nom_bib
```

```python
m_bib = UnsupervisedPCFG(non_terminal_count)
m_bib.fit(corpus_bib, vocab)
```

```python
joblib.dump(m_bib, os.path.join(SRC_DIR, 'results/models/m_bib.joblib'))
```
